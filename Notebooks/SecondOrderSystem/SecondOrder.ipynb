```python
import numpy as np
import sympy as sp
import scienceplots 
import matplotlib.pyplot as plt

plt.style.use(["science","no-latex"])
```

```python
from sympy.abc import w,t,zeta,s
```

```python
f=(1-sp.cos(w*t).subs(w,sp.pi))*sp.Heaviside(t)
sp.plot(f,color="black",title="Plot of response of second order undamped system to a step input",xlim=[0,10])
plt.show()
```

```python
TF=w**2/(s**2+2*zeta*w*s+w**2)
TF
```

```python
R_s=1/s
R_s
```

```python
C_s=R_s*TF
C_s
```

```python
C_t=sp.integrals.inverse_laplace_transform(C_s,s,t)
C_t
```

```python
sp.plot(C_t.subs({zeta:0.1,w:1}),xlim=[0,20],title="Response to step input when $\zeta=0.5$")
plt.show()
```
